```python
%matplotlib inline
import time
import lasio
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

# ── PARÁMETROS AJUSTABLES ─────────────────────────────
LAS_DIR       = Path("./data/cuencas_force-norway")
OUTPUT_DIR    = Path("./output_master/pairplots")
MAX_FILES     = 10
NULL_VALUE    = -9999.25
MIN_VARIABLES = 2
NAN_THRESH    = 0.95

print(f"Archivos encontrados: {len(list(LAS_DIR.glob('*.las')))}")
```

```python
def load_las_files(las_dir: Path, max_files: int, null_value: float) -> dict[str, pd.DataFrame]:
    """
    Lee hasta max_files archivos .las.
    Retorna: {nombre_stem: DataFrame crudo con DEPTH como columna}
    """
    las_files = sorted(las_dir.glob("*.las"))[:max_files]

    if not las_files:
        raise FileNotFoundError(f"No se encontraron archivos .las en: {las_dir}")

    raw_dfs = {}
    print(f"{'Archivo':<35} {'Filas':>8} {'Columnas':>10}")
    print("-" * 55)

    for path in las_files:
        try:
            las = lasio.read(str(path))
            df = las.df().reset_index()
            df.replace(null_value, np.nan, inplace=True)
            raw_dfs[path.stem] = df
            print(f"✅ {path.name:<33} {df.shape[0]:>8} {df.shape[1]:>10}")
        except Exception as e:
            print(f"❌ {path.name}: {e}")

    if not raw_dfs:
        raise ValueError("No se pudo cargar ningún archivo .las válido")

    print(f"\nTotal cargados: {len(raw_dfs)} / {len(las_files)} archivos")
    return raw_dfs
```

```python
def qc_physical_ranges(
    raw_dfs: dict[str, pd.DataFrame],
    ranges: dict[str, tuple]
) -> dict[str, pd.DataFrame]:
    """
    Reemplaza con NaN los valores fuera de rangos físicos.
    Emite un reporte de cuántos valores fueron invalidados por curva.
    """
    qc_dfs = {}
    print(f"\n{'Archivo':<35} {'Curva':<12} {'Inválidos':>10} {'% del total':>12}")
    print("-" * 72)

    for name, df in raw_dfs.items():
        df_qc = df.copy()
        for col, (lo, hi) in ranges.items():
            if col not in df_qc.columns:
                continue
            mask = (df_qc[col] < lo) | (df_qc[col] > hi)
            n_invalid = mask.sum()
            if n_invalid > 0:
                pct = n_invalid / len(df_qc) * 100
                print(f"{name:<35} {col:<12} {n_invalid:>10} {pct:>11.1f}%")
                df_qc.loc[mask, col] = np.nan  # invalida, no elimina
        qc_dfs[name] = df_qc

    return qc_dfs
```

```python
def filter_and_standardize(
    raw_dfs: dict[str, pd.DataFrame],
    min_vars: int,
    nan_thresh: float
) -> tuple[dict[str, pd.DataFrame], list[str]]:
    """
    Etapas:
      1. Descarta DataFrames con menos de min_vars columnas numéricas.
      2. Elimina columnas con proporción de NaN > nan_thresh.
      3. Calcula la intersección de columnas comunes a todos los DFs.
      4. Recorta cada DF a esas columnas y elimina filas con NaN residuales.
      5. Emite reporte de consistencia de filas.
    """
    filtered = {}
    for name, df in raw_dfs.items():
        numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
        if len(numeric_cols) >= min_vars:
            filtered[name] = df[numeric_cols]
        else:
            print(f"⚠️  '{name}' descartado — solo {len(numeric_cols)} variable(s) numérica(s).")

    if not filtered:
        raise ValueError("Ningún archivo pasó el filtro mínimo de variables.")

    cleaned = {}
    for name, df in filtered.items():
        nan_ratio = df.isna().mean()
        cols_keep = nan_ratio[nan_ratio <= nan_thresh].index.tolist()
        cols_dropped = nan_ratio[nan_ratio > nan_thresh].index.tolist()

        if cols_dropped:
            print(f"🗑️  '{name}' — columnas eliminadas (>{int(nan_thresh*100)}% NaN): {cols_dropped}")

        cleaned[name] = df[cols_keep]

    common_cols = set(cleaned[next(iter(cleaned))].columns)
    for df in cleaned.values():
        common_cols &= set(df.columns)
    common_cols = sorted(common_cols)

    if len(common_cols) < 2:
        raise ValueError(
            f"Solo {len(common_cols)} columna(s) en común entre todos los archivos.\n"
            "Considera reducir MIN_VARIABLES o revisar los mnemonics de tus curvas."
        )

    print(f"\n🔍 Columnas comunes detectadas ({len(common_cols)}): {common_cols}")

    standardized = {}
    print(f"\n{'Archivo':<35} {'Filas brutas':>13} {'Filas con NaN':>14} {'Filas finales':>14}")
    print("-" * 78)

    for name, df in cleaned.items():
        df_common = df[common_cols]
        rows_raw = len(df_common)
        nan_rows = int(df_common.isna().any(axis=1).sum())
        df_clean = df_common.dropna().reset_index(drop=True)
        rows_final = len(df_clean)

        print(f"{name:<35} {rows_raw:>13} {nan_rows:>14} {rows_final:>14}")

        if rows_final == 0:
            print(f"   ❌ '{name}' sin filas válidas tras limpieza — omitido.")
            continue

        standardized[name] = df_clean

    print(f"\nArchivos listos para graficar: {len(standardized)}")
    return standardized, common_cols
```

```python
def plot_pairplot(std_dfs: dict[str, pd.DataFrame], output_dir: Path) -> None:
    """
    Genera, muestra dentro del notebook y guarda un pairplot
    por cada DataFrame estandarizado.
    """
    output_dir.mkdir(parents=True, exist_ok=True)
    total = len(std_dfs)

    for i, (name, df) in enumerate(std_dfs.items(), start=1):
        n_vars = df.shape[1]
        n_rows = df.shape[0]

        print(f"\n[{i}/{total}] 📊 Generando pairplot: '{name}'")
        print(f"         {n_rows} muestras × {n_vars} variables")

        g = sns.pairplot(
            df,
            diag_kind="kde",
            kind="scatter",
            plot_kws={"alpha": 0.35, "s": 8, "edgecolors": "none"},
            diag_kws={"fill": True},
            height=2.0,
            aspect=1.0,
            corner=False,
        )

        g.figure.suptitle(
            f"{name}  |  {n_rows} muestras  |  {n_vars} variables",
            y=1.02, fontsize=11, fontweight="bold"
        )

        save_path = output_dir / f"pairplot_{name}.png"
        g.figure.savefig(save_path, dpi=150, bbox_inches="tight")
        print(f"         💾 Guardado: {save_path}")

        plt.show()
        plt.close("all")

    print(f"\n✅ {total} pairplots generados en: {output_dir}")
```

```python
def plot_global_pairplot_and_corr(
    std_dfs: dict[str, pd.DataFrame],
    common_cols: list[str],
    output_dir: Path
) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Genera:
    1) Un pairplot global con todos los registros/pozos en un solo gráfico,
       coloreados por archivo.
    2) Un pairplot global combinado, sin colores, usando todos los datos juntos.
    3) Una matriz de correlación global (-1 a 1) usando todos los datos combinados.

    Retorna:
      - global_df: DataFrame combinado con columna WELL
      - corr: matriz de correlación global
    """
    output_dir.mkdir(parents=True, exist_ok=True)

    combined = []
    for name, df in std_dfs.items():
        temp = df.copy()
        temp["WELL"] = name
        combined.append(temp)

    global_df = pd.concat(combined, ignore_index=True)

    plot_cols = [c for c in common_cols if c in global_df.columns]
    global_df = global_df[plot_cols + ["WELL"]]

    print(f"\nDataFrame global combinado: {global_df.shape[0]} filas, {global_df.shape[1]} columnas")
    print(f"Pozos/archivos incluidos: {global_df['WELL'].nunique()} -> {sorted(global_df['WELL'].unique())}")

    print("\nGenerando pairplot global por archivo...")
    g1 = sns.pairplot(
        global_df,
        vars=plot_cols,
        hue="WELL",
        diag_kind="kde",
        corner=False,
        plot_kws={"alpha": 0.45, "s": 10, "edgecolor": "none"},
        diag_kws={"fill": True},
        height=2.0,
        aspect=1.0
    )

    g1.figure.suptitle(
        f"Pairplot global por archivo - {global_df['WELL'].nunique()} pozos",
        y=1.02,
        fontsize=12,
        fontweight="bold"
    )

    pairplot_path1 = output_dir / "pairplot_global_por_archivo.png"
    g1.figure.savefig(pairplot_path1, dpi=180, bbox_inches="tight")
    plt.show()
    plt.close("all")
    print(f"Guardado: {pairplot_path1}")

    print("\nGenerando pairplot global combinado...")
    global_combined = global_df[plot_cols].copy()

    g2 = sns.pairplot(
        global_combined,
        diag_kind="kde",
        corner=False,
        plot_kws={"alpha": 0.30, "s": 8, "edgecolor": "none", "color": "steelblue"},
        diag_kws={"fill": True, "color": "steelblue"},
        height=2.0,
        aspect=1.0
    )

    g2.figure.suptitle(
        f"Pairplot global combinado - {global_combined.shape[0]} muestras totales",
        y=1.02,
        fontsize=12,
        fontweight="bold"
    )

    pairplot_path2 = output_dir / "pairplot_global_combinado.png"
    g2.figure.savefig(pairplot_path2, dpi=180, bbox_inches="tight")
    plt.show()
    plt.close("all")
    print(f"Guardado: {pairplot_path2}")

    print("\nGenerando matriz de correlación global...")
    corr = global_combined.corr(method="pearson")

    plt.figure(figsize=(10, 8))
    sns.heatmap(
        corr,
        annot=True,
        fmt=".2f",
        cmap="coolwarm",
        vmin=-1,
        vmax=1,
        center=0,
        square=True,
        linewidths=0.5,
        cbar_kws={"label": "Correlación"}
    )
    plt.title("Matriz de correlación global combinada", fontsize=13, fontweight="bold")
    plt.xticks(rotation=45, ha="right")
    plt.yticks(rotation=0)
    plt.tight_layout()

    corr_path = output_dir / "matriz_correlacion_global.png"
    plt.savefig(corr_path, dpi=180, bbox_inches="tight")
    plt.show()
    plt.close()
    print(f"Guardado: {corr_path}")

    return global_df, corr
```

```python
def format_engineering_stats(stats_df: pd.DataFrame) -> pd.DataFrame:
    """
    Toma la salida de describe().T y devuelve una sola tabla
    con formato ingenieril.
    """
    stats = stats_df.copy()

    stats = stats.rename(columns={
        "count": "N",
        "mean": "Mean",
        "50%": "P50",
        "std": "Std",
        "min": "Min",
        "25%": "P25",
        "75%": "P75",
        "max": "Max",
    })

    stats["IQR"] = stats["P75"] - stats["P25"]
    stats["CV"] = stats["Std"] / stats["Mean"].replace(0, np.nan)

    cols = ["CURVE", "N", "Mean", "P50", "Std", "Min", "P25", "P75", "Max", "IQR", "CV"]
    return stats.loc[:, cols]


def compute_stats_per_well(
    std_dfs: dict[str, pd.DataFrame],
    output_dir: Path
) -> pd.DataFrame:
    """
    Genera una sola tabla de estadísticos descriptivos por pozo
    en formato ingenieril.
    Guarda un CSV individual por pozo y devuelve un DataFrame consolidado.
    """
    output_dir.mkdir(parents=True, exist_ok=True)
    all_stats = []

    for name, df in std_dfs.items():
        num_df = df.select_dtypes(include="number")

        if num_df.empty:
            print(f"⚠️  '{name}' no tiene columnas numéricas válidas para estadística.")
            continue

        stats = num_df.describe().T
        stats.index.name = "CURVE"
        stats.reset_index(inplace=True)

        stats = format_engineering_stats(stats)
        stats["WELL"] = name
        stats = stats[["WELL", "CURVE", "N", "Mean", "P50", "Std", "Min", "P25", "P75", "Max", "IQR", "CV"]]

        all_stats.append(stats)

        csv_path = output_dir / f"stats_{name}.csv"
        stats.to_csv(csv_path, index=False)
        print(f"Tabla estadística individual guardada: {csv_path}")

    if not all_stats:
        raise ValueError("No se generaron tablas estadísticas por pozo.")

    stats_all_wells = pd.concat(all_stats, ignore_index=True)
    return stats_all_wells


def build_global_df(
    std_dfs: dict[str, pd.DataFrame],
    common_cols: list[str]
) -> pd.DataFrame:
    """
    Combina todos los DataFrames estandarizados en uno solo,
    añadiendo una columna WELL con el nombre del archivo.
    """
    combined = []

    for name, df in std_dfs.items():
        temp = df.copy()
        temp["WELL"] = name
        combined.append(temp)

    if not combined:
        raise ValueError("No hay DataFrames estandarizados para combinar.")

    global_df = pd.concat(combined, ignore_index=True)
    plot_cols = [c for c in common_cols if c in global_df.columns]

    return global_df[plot_cols + ["WELL"]]


def compute_global_stats(
    std_dfs: dict[str, pd.DataFrame],
    common_cols: list[str],
    output_dir: Path
) -> pd.DataFrame:
    """
    Genera una sola tabla de estadísticos globales con formato ingenieril
    usando todos los registros combinados.
    Devuelve el DataFrame y lo guarda como CSV.
    """
    output_dir.mkdir(parents=True, exist_ok=True)

    global_df = build_global_df(std_dfs, common_cols)
    num_df = global_df.select_dtypes(include="number")

    if num_df.empty:
        raise ValueError("El DataFrame global no contiene columnas numéricas válidas.")

    global_stats = num_df.describe().T
    global_stats.index.name = "CURVE"
    global_stats.reset_index(inplace=True)

    global_stats = format_engineering_stats(global_stats)

    csv_path = output_dir / "stats_global_todos_registros.csv"
    global_stats.to_csv(csv_path, index=False)
    print(f"Tabla estadística global guardada: {csv_path}")

    return global_stats
```

```python
# ══════════════════════════════════════════════════
#  CELDA 6 — PIPELINE PRINCIPAL (ejecuta todo)
# ══════════════════════════════════════════════════
t0_total = time.perf_counter()

# ETAPA 1: Carga
raw_dfs = load_las_files(
    las_dir    = LAS_DIR,
    max_files  = MAX_FILES,
    null_value = NULL_VALUE
)

PHYSICAL_RANGES = {
    "GR":   (0, 300), "RHOB": (1.0, 3.2),
    "NPHI": (-0.15, 1.0), "DTC": (40, 240),
    "RDEP": (0.01, 50000), "RMED": (0.01, 50000), "CALI": (4, 30),
}
raw_dfs = qc_physical_ranges(raw_dfs, PHYSICAL_RANGES)

# ETAPA 2: Filtrado y estandarización
std_dfs, common_cols = filter_and_standardize(
    raw_dfs    = raw_dfs,
    min_vars   = MIN_VARIABLES,
    nan_thresh = NAN_THRESH
)

# ETAPA 2.5: Transformación log10 para resistividades (solo visualización)
RESISTIVITY_COLS = ["RDEP", "RMED"]
std_dfs_plot = {}
for name, df in std_dfs.items():
    df_plot = df.copy()
    for col in RESISTIVITY_COLS:
        if col in df_plot.columns:
            df_plot[f"log10_{col}"] = np.where(df_plot[col] > 0, np.log10(df_plot[col]), np.nan)
            df_plot.drop(columns=[col], inplace=True)
    std_dfs_plot[name] = df_plot

common_cols_plot = [f"log10_{c}" if c in RESISTIVITY_COLS else c for c in common_cols]

# ETAPA 3: Visualización individual (con log10)
plot_pairplot(std_dfs=std_dfs_plot, output_dir=OUTPUT_DIR)

# ETAPA 4: Pairplot global + correlación (con log10)
global_df, corr_global = plot_global_pairplot_and_corr(
    std_dfs=std_dfs_plot,
    common_cols=common_cols_plot,
    output_dir=OUTPUT_DIR
)

# ETAPA 5: Estadísticas (datos originales en ohm·m)
stats_per_well = compute_stats_per_well(
    std_dfs=std_dfs,
    output_dir=OUTPUT_DIR
)
stats_global = compute_global_stats(
    std_dfs=std_dfs,
    common_cols=common_cols,
    output_dir=OUTPUT_DIR
)
display(stats_global)

# ETAPA 6: Exportar CSV y NPY (datos originales en ohm·m)
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
print(f"\n{'Archivo':<35} {'Filas':>8} {'Cols':>5}  Archivos generados")
print("-" * 80)

for name, df in std_dfs.items():
    csv_path = OUTPUT_DIR / f"{name}.csv"
    npy_path = OUTPUT_DIR / f"{name}.npy"
    df.to_csv(csv_path, index=False)
    np.save(str(npy_path), df.values.astype(np.float64))
    print(f"{name:<35} {df.shape[0]:>8} {df.shape[1]:>5}  {csv_path.name} | {npy_path.name}")

all_dfs = []
for name, df in std_dfs.items():
    temp = df.copy()
    temp["WELL"] = name
    all_dfs.append(temp)

global_export_df = pd.concat(all_dfs, ignore_index=True)
csv_global = OUTPUT_DIR / "ALL_WELLS.csv"
npy_global = OUTPUT_DIR / "ALL_WELLS.npy"
global_export_df.to_csv(csv_global, index=False)
np.save(str(npy_global), global_export_df.select_dtypes(include=np.number).values.astype(np.float64))

print("-" * 80)
print(f"{'ALL_WELLS (global)':<35} {global_export_df.shape[0]:>8} {global_export_df.shape[1]:>5}  {csv_global.name} | {npy_global.name}")
print(f"\n✅ Exportados {len(std_dfs)} individuales + 1 global → {OUTPUT_DIR}")

t1_total = time.perf_counter()
print(f"\n⏱️ Tiempo total de ejecución: {t1_total - t0_total:.2f} segundos")
```
